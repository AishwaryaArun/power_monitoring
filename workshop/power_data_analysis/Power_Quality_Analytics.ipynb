```python
# Count number of power cuts
# generates -1 for Power cut, +1 for Power restored, else 0
#power cut transition is -1. 
#power restored transition is +1
#Power staus unchanged is 0

# This is done by subtracting present datapoint and the previous data point( shift() function takes care of it)
# When shifted, 1st values has NaN. Fillna(0) will fill 0 for values having NaN
df['delta'] = (df['on_off']-df['on_off'].shift()).fillna(0)
```

```python
# computing duration between any two events
df_OnOff['delta_time'] = (df_OnOff['unix_conv']-df_OnOff['unix_conv'].shift()).fillna(0)
df_OnOff
```

```python
# calculating total ON and OFF duration
on=df_OnOff.unix_conv[0]-df_OnOff.unix_conv[0]
off=df_OnOff.unix_conv[0]-df_OnOff.unix_conv[0]

for i in range (len(df_OnOff)-1):
				 if df_OnOff.on_off[i+1]==0 and df_OnOff.on_off[i]==1 :
				     on=on+df_OnOff.delta_time[i+1]
				 elif df_OnOff.on_off[i+1]==1 and df_OnOff.on_off[i]==1 :
				     on=on+df_OnOff.delta_time[i+1]
				 elif df_OnOff.on_off[i+1]==0 and df_OnOff.on_off[i]==0 :
				     off=off+df_OnOff.delta_time[i+1]
				 else :
				     off=off+df_OnOff.delta_time[i+1]
```
